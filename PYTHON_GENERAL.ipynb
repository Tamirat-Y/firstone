# IF ----ELSE
Non-Boolean Conditions
Note that conditions do not necessarily have to be booleans. In fact, a condition can be any value. The value is converted into a boolean automatically using the bool operator. This means that falsy values like 0, '', {}, [],None etc. evaluate to False and all other values evaluate to True.

```python
if '':
    print('The condition evaluted to True')
else:
    print('The condition evaluted to False')
```

Python provides a shorter syntax, which allows writing such conditions in a single line of code. It is known as a conditional expression, sometimes also referred to as a ternary operator. It has the following syntax:

x = true_value if condition else false_value

```python
a_number=3
parity = 'even' if a_number % 2 == 0 else 'odd'
print(parity)
```

# ITERATION

range(a, b) - Creates a sequence of numbers from a to b-1

range(a, b, step) - Creates a sequence of numbers from a to b-1 with increments of step

Another way to achieve the same result is by using the enumerate function with a_list as an input, which returns a tuple containing the index and the corresponding element.

```python
a_list = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday']

for i in range(len(a_list)):
    print('The value at position {} is {}.'.format(i, a_list[i]))
```

```python
#change the number of days and select the one woth low cost cities
def needd(a):
    total=a[0]+a[1]*7+a[2]
    return {a[3]:total}
y={} 

ar=[[200,20,200,'paris'],[250,30,120,'lodon'],[370,15,80,'dubai'],[450,10,70,'mumbai']]

for i in ar:
    y.update(needd(i))

print('the cost- city table is:\n',y)

costs=list(y.values())
citys=list(y.keys())
l=min(costs)

l_cost=[]
for i in range(len(costs)):
    if costs[i]==l:
        l_cost.append(i)
name=''
z=0
for x in l_cost:
    if z==0:
        name=name+' '+citys[x]
    else:
         name=name+', '+citys[x]
    z=z+1

print('the list cost cities are/ is {} with cost {}'.format(name,l))
```

```python
#minimum days
def needd(a):
    total_cost=0
    days=1
    while True:  
        total=a[0]+a[1]*days+a[2]
        if total<=1000:
            total_cost=total
            days=days+1
        else:
            break
   
    return {a[3]:[days-1, total_cost]}


y={}

ar=[[200,20,200,'paris'],[250,30,120,'lodon'],[370,15,80,'dubai'],[450,10,70,'mumbai']]

for k in ar:
     y.update(needd(k))
   
print('the cost- city table is:\n',y)

citys=list(y.keys())
days=[]
costs=[]
for j in citys:
        days=days+ [y[j][0],]
        costs=costs+ [y[j][1],]

min_days=min(days)

min_days_index=[]
for i in range(len(days)):
    if days[i]==min_days:
      min_days_index.append(i)
name=''
z=0
for x in min_days_index:
    if z==0:
        name=name+' '+citys[x]
    else:
         name=name+', '+citys[x]
    z=z+1

print('with the given cost the cities are / is {} with number of days {}'.format(name,min_days))
```
